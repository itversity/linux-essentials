

## Startup Levels and Scripts Overview

Let us go through the details related to Startup Levels and Scripts.

The third party services will be enabled at level 7 from (0-6). 0 level states unless all the services which are specified as priority state as 0 are stated your system should not be considered in the healthy state. As part of the troubleshooting process system admins may restart the services using <init 0> to shut down. If you want to restart the complete system you use <init 6>.
